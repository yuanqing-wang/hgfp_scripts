```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
gcn_rmse_tr = np.load('gcn_qm9/res/rmse_tr.npy')
gcn_rmse_vl = np.load('gcn_qm9/res/rmse_vl.npy')
gcn_cr_rmse_tr = np.load('gcn_with_combine_readout_qm9/res/rmse_tr.npy')
gcn_cr_rmse_vl = np.load('gcn_with_combine_readout_qm9/res/rmse_vl.npy')
```

```python
plt.style.use('fivethirtyeight')
```

```python
plt.figure(facecolor='white')
plt.plot(gcn_cr_rmse_vl, label=r'$RMSE_\mathtt{VALIDATION, GFP}$', alpha=0.8)
plt.plot(gcn_cr_rmse_tr, label=r'$RMSE_\mathtt{TRAINING, GFP}$', alpha=0.8)
plt.plot(gcn_rmse_tr, label=r'$RMSE_\mathtt{TRAINING, GCN}$', alpha=0.8)
plt.plot(gcn_rmse_vl, label=r'$RMSE_\mathtt{VALIDATION, GCN}$', alpha=0.8)
plt.legend()
plt.xlabel(r'$N_\mathtt{epochs}$')
plt.ylabel('kcal/mol')
plt.tight_layout()
plt.savefig('rmse_compare.png', dpi=500, facecolor='white')
# 
```

```python
u = np.load('gcn_with_combine_readout_qm9/res/u_te.npy')
u_hat = np.load('gcn_with_combine_readout_qm9/res/u_hat_te.npy')
```

```python
plt.style.use('fivethirtyeight')
fig = plt.figure()
fig.set_facecolor('white')
# plt.axis('equal')
plt.gca().set_aspect('equal', adjustable='box')
plt.plot(np.linspace(-2000, -500, 10), np.linspace(-2000, -500, 10), c='k', linewidth=0.5)
plt.scatter(u, u_hat, s=5, alpha=0.5)

plt.yticks([-2000, -1500, -1000, -500])
plt.xticks([-2000, -1500, -1000, -500])
plt.xlim((-2000, -500))
plt.ylim((-2000, -500))

plt.xlabel('$U$ (kcal/mol)')
plt.ylabel(r'$\hat{U}$ (kcal/mol)')
plt.tight_layout()
plt.savefig('scatter.png', dpi=500, facecolor='white')
```

```python
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(8, 6))
fig.set_facecolor('white')

bond_k_true = np.load('gcn_param/res/bond_k_y_te.npy')
bond_k_pred = np.load('gcn_param/res/bond_k_y_hat_te.npy')
angle_k_true = np.load('gcn_param/res/angle_k_y_te.npy')
angle_k_pred = np.load('gcn_param/res/angle_k_y_hat_te.npy')
atom_k_true = np.load('gcn_param/res/atom_k_y_te.npy')
atom_k_pred = np.load('gcn_param/res/atom_k_y_hat_te.npy')

bond_eq_true = np.load('gcn_param/res/bond_eq_y_te.npy')
bond_eq_pred = np.load('gcn_param/res/bond_eq_y_hat_te.npy')
angle_eq_true = np.load('gcn_param/res/angle_eq_y_te.npy')
angle_eq_pred = np.load('gcn_param/res/angle_eq_y_hat_te.npy')
atom_eq_true = np.load('gcn_param/res/atom_eq_y_te.npy')
atom_eq_pred = np.load('gcn_param/res/atom_eq_y_hat_te.npy')






ax0 = plt.subplot(2, 3, 1)
ax0.set_aspect('equal')
ax0.scatter(bond_k_true, bond_k_pred, s=5, alpha=0.5)
ax0.set_xlim(500, 1200)
ax0.set_ylim(500, 1200)
ax0.set_xticks((750, 1000))
ax0.set_yticks((750, 1000))
ax0.set_title(r'$k_\mathtt{bond}$')

ax1 = plt.subplot(2, 3, 2)
ax1.set_aspect('equal')
ax1.scatter(angle_k_true, angle_k_pred, s=5, alpha=0.5)
ax1.set_xlim(0, 800)
ax1.set_ylim(0, 800)
ax1.set_xticks((200, 400, 600))
ax1.set_yticks((200, 400, 600))
ax1.set_title(r'$k_\mathtt{angle}$')

ax2 = plt.subplot(2, 3, 3)
ax2.set_aspect('equal')
ax2.scatter(atom_k_true, atom_k_pred, s=5, alpha=0.5)
ax2.set_xlim(0, 0.25)
ax2.set_ylim(0, 0.255)
ax2.set_title('$\epsilon$')

ax3 = plt.subplot(2, 3, 4)
ax3.set_aspect('equal')
ax3.scatter(bond_eq_true, bond_eq_pred, s=5, alpha=0.5)
ax3.set_xlim(1.2, 1.7)
ax3.set_ylim(1.2, 1.7)
ax3.set_title(r'$l_\mathtt{eq, bond}$')

ax4 = plt.subplot(2, 3, 5)
ax4.set_aspect('equal')
ax4.scatter(angle_eq_true, angle_eq_pred, s=5, alpha=0.5)
ax4.set_xlim(0., 200.)
ax4.set_ylim(0., 200.)
ax4.set_xticks((50, 100, 150))
ax4.set_yticks((50, 100, 150))
ax4.set_title(r'$\phi_\mathtt{eq, angle}$')

ax5 = plt.subplot(2, 3, 6)
ax5.set_aspect('equal')
ax5.scatter(atom_eq_true, atom_eq_pred, s=5, alpha=0.5)
ax5.set_xlim(1.5, 2)
ax5.set_ylim(1.5, 2)
ax5.set_title('$\sigma$')

for ax in [ax0, ax1, ax2, ax3, ax4, ax5]:
    ax.plot(np.linspace(-2000, 2000), np.linspace(-2000, 2000), linewidth=1, c='k')

plt.tight_layout()
plt.savefig('param.png', dpi=500, facecolor='white')




```
